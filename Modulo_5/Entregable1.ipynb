```python
# Selecciono únicamente los sectores Retail Trade y Wholesale Trade
df_main = df_main[df_main['Sector NAICS\nlevel 1'].str.contains('Retail Trade|Wholesale Trade')]
```

```python
df_main['R'] = np.log(df_main.groupby(['firm'])['adjprice'].shift(-1)) - np.log(df_main.groupby(['firm'])['adjprice'].shift(3))

df_main.head(1000)
df_main.to_csv("prueba1.csv")

df_q1 = df_main[df_main['q'] == '2022q1']

df_q1 = df_q1[['firm','q','revenue','cogs','sgae','otheropexp','finexp','incometax',
              'totalassets','totalliabilities','adjprice','originalprice','sharesoutstanding',
              'Sector NAICS\nlevel 1','R']]
```

```python
df_q1.head()
```

```python
# Agregamos las columnas de las variables extras haciendo uso de las ecuaciones que vimos en clase

# Formula de EBIT
df_q1['EBIT'] = df_q1['revenue'] - df_q1['cogs'] - df_q1['sgae'] - df_q1['otheropexp']
# Formula de Net Income
df_q1['NetIncome'] = df_q1['EBIT'] - df_q1['incometax'] - df_q1['finexp']
# Formula de EPS
df_q1['EPS'] = df_q1['NetIncome'] / df_q1['sharesoutstanding']
# Formula de Book
df_q1['Book'] = df_q1['totalassets'] - df_q1['totalliabilities']
# Formula de Market
df_q1['Market'] = df_q1['originalprice'] * df_q1['sharesoutstanding']
# Formula de Book to Market Ratio
df_q1['BTMR'] = df_q1['Book'] / df_q1['Market']
# Formula de EPSP
df_q1['EPSP'] = df_q1['EPS'] / df_q1['originalprice']
# Formuña de OPM
df_q1['OPM'] = df_q1['EBIT'] / df_q1['revenue']
```

```python

df_q1 = df_q1.sort_values(by = 'Market').reset_index(drop=True,inplace = True)
```
